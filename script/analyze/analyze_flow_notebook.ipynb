```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

from datetime import time, datetime
```

```python
def filter_warming_up_period(df, time):
    return df[df.TIME.dt.time >= time]

df_flow = pd.read_csv('flow.csv', sep=';')
df_flow['TIME'] = pd.to_datetime(df_flow['TIME'])
df_flow = filter_warming_up_period(df_flow, time(hour=6, minute=45))
df_flow.head()
```

```python
for mode in df_flow.VEHICLE_TYPE.unique():
    df_flow_mode = df_flow[df_flow.VEHICLE_TYPE == mode]

    fig, axes = plt.subplots(1, 2, figsize=(10,4))
    df_flow_mode.plot(ax=axes[0], x='TIME', y='ACCUMULATION')
    df_flow_mode.plot(ax=axes[1], x='TIME', y='SPEED')
    plt.show()
```

```python
def plot_speed_accumulation(df_flow):
    for df_mode in df_flow.groupby('VEHICLE_TYPE'):
        fig, ax = plt.subplots(1,2, figsize=(10,4))
        df_mode[1].plot(x='TIME', y='ACCUMULATION', ax=ax[0], title=df_mode[0])
        df_mode[1].plot(x='TIME', y='SPEED', ax=ax[1], title=df_mode[0])
        plt.show()
```

```python
plot_speed_accumulation(df_flow)
```
