```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# User file path
file_path = "user.csv"
df = pd.read_csv('user.csv', sep = ';')

# Conversion des donn√©es temporelles 
df['TIME'] = pd.to_datetime(df['TIME'])
df.head()
```

```python
df['STATE'].value_counts()
```

```python
df.ID.nunique()
```

```python
# Creation du dataframe des users
s_nb_vehicles = df.groupby('ID').VEHICLE.nunique()#.describe()
s_travel_time = df.groupby('ID').last()['TIME']-df.groupby('ID').first()['TIME']
s_arrived = df.groupby('ID').STATE.last() == 'ARRIVED'
s_distance = df.groupby('ID').DISTANCE.last()

df_users = pd.concat([s_nb_vehicles, s_travel_time, s_arrived, s_distance],
                    keys = ['NB_VEHICLES', 'TIME', 'ARRIVED', 'DISTANCE'], axis=1)
df_users
```

```python
df_users['TIME_min'] = df_users['TIME'].dt.total_seconds()/60
```

```python
df_users
```

```python
sns.histplot(data=df_users, x='DISTANCE', bins=15)
```

```python
sns.histplot(data=df_users, x='TIME_min', bins=15)
```
